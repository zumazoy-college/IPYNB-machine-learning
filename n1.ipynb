<p>Импорт необходимых библиотек (pandas - для работы с табличными данными)</p>

```python
import pandas as pd
```

<strong>1. Загрузка датасета.</strong>
<p>pd.read_csv() - функция для чтения CSV файлов</p>

```python
path = r"C:\Users\Zumazoy\Datasets\youtube-top-100-songs-2025.csv"
df = pd.read_csv(path)
```

<strong>2. Просмотр и исследование датасета.</strong>
<p>df.info() - показывает информацию о датафрейме: типы данных, количество ненулевых значений и т.д.</p>

```python
df.info()
```

<p>df.shape - возвращает кортеж (количество_строк, количество_столбцов)</p>

```python
print(f"Количество строк: {df.shape[0]}. Количество столбцов: {df.shape[1]}.")
```

<p>df.head() - показывает первые 5 строк датафрейма по умолчанию</p>

```python
df.head()
```

<p>df.describe() - показывает статистику для числовых столбцов: count, mean, std, min, max и др.</p>

```python
df.describe()
```

<p>df.isnull().sum() - подсчитывает количество пропущенных значений в каждом столбце</p>

```python
df.isnull().sum()
```

<p>df['column'].value_counts() - подсчитывает количество каждого уникального значения в столбце</p>

```python
print(f"Топ 5 самых популярных каналов:\n{df['channel'].value_counts().head()}")
```

<p>df['column'].nunique() - возвращает количество уникальных значений в столбце</p>

```python
print(f"Количество уникальных каналов: {df['channel'].nunique()}")
```

<strong>3. Удаление признака.</strong>
<p>df.drop('column', axis=1) - удаляет указанный столбец из датафрейма, axis=1 означает, что мы работаем со столбцами, а не со строками</p>

```python
df = df.drop('fulltitle', axis=1)
print(f"Количество столбцов: {df.shape[1]}")
```

<strong>4. Замена и удаление значений.</strong>
<p>df['column'].replace - заменяет определенные значения в столбце на указаннные</p>

```python
df['categories'] = df['categories'].replace({'People & Blogs': 'Vlog'})
print(f"Значения в 'categories' заменены:\n{df['categories'].value_counts()}")
```

<p>df[df['column'] >= 100000] - удаляет все строчки, где значение меньше 100000</p>

```python
df = df[df['view_count'] >= 100000]
print(f"Количество строк после удаления: {df.shape[0]}")
```

<strong>5. Сортировка по указанному столбцу.</strong>
<p>df.sort_values(by='column', ascending=False) - сортирует датафрейм по указанному столбцу, ascending=False означает сортировку по убыванию</p>

```python
df = df.sort_values(by='view_count', ascending=False)
df[['title', 'channel', 'view_count']].head()
```

<strong>6. Удаление дубликатов.</strong>
<p>df.drop_duplicates() - удаляет дублирующиеся строки, subset=['col1', 'col2'] - указывает, по каким столбцам проверять дубликаты</p>

```python
df = df.drop_duplicates(subset=['title', 'channel'])
print(f"Количество строк после удаления дубликатов: {df.shape[0]}")
```

<strong>7. Выбор целевой переменной.</strong>
<p>Целевая переменная: categories</p>
<p>Потому что:</p>
<ul>
    <li>Категория видео - важный фактор, влияющий на популярность</li>
    <li>Имеет четкую категоризацию (Music vs Vlog)</li>
    <li>Представляет практический интерес для анализа контента</li>
</ul>

<strong>8. Сохранение датасета.</strong>
<p>df.to_csv() - сохраняет датафрейм в CSV файл, index=False - не сохранять индексы строк, encoding='utf-8' - кодировка для поддержки специальных символов</p>

```python
df.to_csv('youtube_top_songs_2025.csv', index=False, encoding='utf-8')
```

<h5>Вывод: удален признак 'fulltitle', произведена замена значений categories, удаление значений view_count, сортировка по просмотрам видео и удаление дубликатов.</h5>
