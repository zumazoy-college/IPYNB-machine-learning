<h5>Ход работы:</h5>

```python
import pandas as pd # для работы с табличными данными и анализа временных рядов
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler, Normalizer, KBinsDiscretizer, PowerTransformer
```

```python
path = r"C:\\Users\\Zumazoy\\Datasets\\youtube-top-100-songs-2025.csv"
df = pd.read_csv(path) # загружает данные из CSV-файла в DataFrame pandas
df.info() # выводит сводную информацию о DataFrame: структуру, типы данных и количество ненулевых значений
```

<h4>1. OrdinalEncoder - преобразует категориальные значения в числовые коды (0, 1, 2, ...). Используется когда есть естественный порядок между категориями или когда порядок не важен, но нужно числовое представление. Этот метод подходит для алгоритмов, которые могут работать с порядковыми данными.</h4>

```python
categorical_columns = ['categories', 'live_status']

ordinal_encoder = OrdinalEncoder()
df[categorical_columns] = ordinal_encoder.fit_transform(df[categorical_columns])
df
```

<h4>2. LabelEncoder - преобразует каждую категорию в уникальное целое число. Используется специально для кодирования целевых переменных (меток) в задачах классификации. И не подходит для признаков, так как может создать ложное представление о порядке категорий.</h4>

```python
label_column = 'categories'

label_encoder = LabelEncoder()
df[f'{label_column}_label'] = label_encoder.fit_transform(df[label_column])
df
```

<h4>3. OneHotEncoder - создает бинарные (0/1) колонки для каждой категории. Используется когда нет естественного порядка между категориями, чтобы избежать ложного представления о порядке. Этот метод увеличивает размерность данных, но сохраняет равнозначность всех категорий.</h4>

```python
onehot_encoder = OneHotEncoder(sparse_output=False, drop='first')

onehot_encoded = onehot_encoder.fit_transform(df[categorical_columns])
onehot_columns = onehot_encoder.get_feature_names_out(categorical_columns)

onehot_df = pd.DataFrame(onehot_encoded, columns=onehot_columns, index=df.index)
df = pd.concat([df, onehot_df], axis=1)

df
```

<h4>4. StandardScaler - преобразует данные к стандартному нормальному распределению (mean=0, std=1). Используется когда алгоритмы чувствительны к масштабу данных (SVM, линейные модели). Этот метод помогает сравнивать признаки с разными единицами измерения.</h4>

```python
numeric_columns = ['view_count', 'duration', 'channel_follower_count']

standard_scaler = StandardScaler()
df[numeric_columns] = standard_scaler.fit_transform(df[numeric_columns])
df
```

<h4>5. MinMaxScaler - масштабирует данные в заданный диапазон (по умолчанию [0, 1]). Используется когда важно сохранить исходное распределение и когда нужны данные в определенном диапазоне (нейронные сети).</h4>

```python
minmax_scaler = MinMaxScaler()

df[numeric_columns] = minmax_scaler.fit_transform(df[numeric_columns])
df
```

<h4>6. Normalizer - нормализует каждый образец (строку) независимо, чтобы его норма (L1, L2) равнялась 1. Используется когда важны относительные величины признаков, а не абсолютные (текстовые данные, рекомендательные системы). Он работает с каждой строкой данных отдельно, а не с колонками.</h4>

```python
normalizer = Normalizer(norm='l2')  # L2 норма (евклидова)

df[numeric_columns] = normalizer.fit_transform(df[numeric_columns])
df
```

<h4>7. KBinsDiscretizer - преобразует непрерывные данные в дискретные интервалы (бины). Этот метод помогает упростить сложные распределения и снизить влияние выбросов. Он полезен для преобразования непрерывных признаков в категориальные.</h4>

```python
discretizer = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='quantile')

df[['view_count_binned', 'duration_binned']] = discretizer.fit_transform(df[['view_count', 'duration']])
df
```

<h4>8. PowerTransformer - применяет степенные преобразования (Yeo-Johnson или Box-Cox) для приведения распределения к нормальному. Этот метод эффективен когда данные имеют сильную асимметрию. Он помогает стабилизировать дисперсию и улучшить работу алгоритмов, требующих нормального распределения.</h4>

```python
power_transformer = PowerTransformer(method='yeo-johnson')

df[['view_count_power', 'channel_follower_count_power']] = power_transformer.fit_transform(df[['view_count', 'channel_follower_count']])
df
```

```python
df.to_csv('youtube_top_songs_2025_enc.csv', index=False, encoding='utf-8')
```

Дополнительные методы KBinsDiscretizer и PowerTransformer продемонстрировали возможности расширенной предобработки для работы со сложными распределениями.</h5>
